**PROBLEM 1**

```python
import re
#open files in readlines 
locusfile = open("LocusGene.tsv", "r")
Locusfile = locusfile.readlines()

germfile = open("Germplasm.tsv", "r")
Germfile = germfile.readlines()

#length of each file
longlocus = len(Locusfile)
longgerm = len(Germfile)

if longlocus == longgerm:
    print("they have the same size")
else:
    print("nope")

#loop necesary to take all the AGI Locus Code from both files
locus = []
germ = []
for i in range(0,longlocus):
    matchObj = re.search(r'AT\dG\d+', Locusfile[i])
    matchObj2 = re.search(r'AT\dG\d+', Germfile[i])
    if matchObj:
        Locus = matchObj.group()
        locus.append(Locus)
        
        Germ = matchObj2.group()
        germ.append(Germ)
        
    else:
        print("dont match")

print(locus)        
print()
print(germ)
print()

#check if they match
if locus == germ:
    print("they are the same")
else:
    print("they dont match")
    
    
```

**PROBLEM 3**

```python
import pymysql.cursors
connection = pymysql.connect(host='localhost',
                             user='root',
                             password='root',
                             db='exam2',
                             charset='utf8mb4',  # note utf8... this is important for unusual characters!
                             cursorclass=pymysql.cursors.DictCursor)


#connection.autocommit = False

try:
    with connection.cursor() as cursor:
#loop to take all the attributes of each file and insert on each tables
        for i in range(1,len(Germfile)):
            filalocus = Locusfile[i].split('\t')
            
            locus = filalocus[0]
            gene = filalocus[1]
            protein_length = filalocus[2]

            sql = "insert into locustable(locus, gene, protein_length) values ('{}','{}','{}')".format(locus, gene, protein_length)
            cursor.execute(sql)                                                                 
            
            filagerm = Germfile[i].split('\t')
            
            germplasm = filagerm[1]
            phenotype = filagerm[2]
            pubmed = filagerm[3]
                                                                         
            sql = "insert into germtable(locus, germplasm, phenotype, pubmed) values ('{}','{}','{}','{}')".format(locus, germplasm, phenotype, pubmed)
            cursor.execute(sql)                                                                 
            

    connection.commit()

finally:
    print("")
    connection.close()    
```
